```python
# Simple moving average
nifty_50_data["SMA_20"] = nifty_50_data["Close"].rolling(20).mean()

# RSI (simple)
delta = nifty_50_data["Close"].diff()
gain = delta.clip(lower=0).rolling(14).mean()
loss = (-delta.clip(upper=0)).rolling(14).mean()
rs = gain / loss
nifty_50_data["RSI"] = 100 - (100 / (1 + rs))
```
